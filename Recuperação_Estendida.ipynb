# Questão 1 (vale 1 ponto)
Suponha que você esteja projetando uma ponte suspensa e precise determinar a deformação máxima que ocorrerá em um determinado ponto da estrutura. A partir de análises e medições preliminares, você sabe que a deformação em um ponto na posição x da ponte pode ser aproximada por uma função $f(x) = 3x^2 - 2x + 5$. Precisamos saber a deformação quando $x$ se aproxima do valor $2$.

Utilizando a Fórmula das diferenças finitas encontre os limites para cinco pontos à esquerda do ponto desejado e cinco pontos à direita do ponto desejado. Depois traçe um gráfico desta função anote o limite no ponto desejado e imprima os limites nos cinco pontos à esquerda e à direita.




## Resposta

```python
def f(x):
    return 3*x**2 - 2*x + 5

# Ponto desejado
x_desired = 2

# Intervalo
h = 0.1

# Pontos à esquerda do ponto desejado
x_left = np.linspace(x_desired - 5*h, x_desired - h, 5)
y_left = f(x_left)

# Pontos à direita do ponto desejado
x_right = np.linspace(x_desired + h, x_desired + 5*h, 5)
y_right = f(x_right)

# Limites
left_limit = (f(x_desired) - f(x_desired - h)) / h
right_limit = (f(x_desired + h) - f(x_desired)) / h

# Plotagem da função
x = np.linspace(x_desired - 1, x_desired + 1, 100)
y = f(x)

plt.plot(x, y, label='f(x)')
plt.scatter(x_desired, f(x_desired), color='red', label='Ponto desejado')
plt.scatter(x_left, y_left, color='green', label='Pontos à esquerda')
plt.scatter(x_right, y_right, color='blue', label='Pontos à direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) com limites')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir os limites
print('Limite à esquerda:', left_limit)
print('Limite à direita:', right_limit)
```

# Exercício de Limites e Descontinuidades

Considere as seguintes funções:

1. $$ f(x) = \tan(x) $$
2. $$ g(x) = \frac{1}{x-5} $$

Para cada função, trace o gráfico no intervalo de 0 a 10 usando Python, Matplotlib e NumPy. a)Determine, se existem, ou não descontinuidade para estas funções no intervalo solicitado. b)Caso exista alguma descontinuidade encontre o limite em um ponto qualquer à direita e outro a esquerda do ponto onde existe a descontinuidade. c)Você precisa plotar os gráficos destas funções usando python, e apenas as bibliotecas matplotlib, numpy e sympy.


## Resposta


Função 1

a- Para a função f(x)=tan(x), existem descontinuidades quando o X é muktiplo de π/2, no intervalo de 0 a 10 elas ororrem em x = π/2, x = 3π/2, x = 5π/2, x = 7π/2 e x = 9π/2

```python
## Função 1

# Função f(x) = tan(x)
x_sp = sp.symbols('x')
f_sp = sp.tan(x_sp)

# Pontos de descontinuidade
discontinuity_points = [np.pi/2, 3*np.pi/2, 5*np.pi/2, 7*np.pi/2, 9*np.pi/2]

# Cálculo dos limites à direita e à esquerda dos pontos de descontinuidade
limits_right = []
limits_left = []

for point in discontinuity_points:
    limit_right = sp.limit(f_sp, x_sp, point, dir='+')
    limit_left = sp.limit(f_sp, x_sp, point, dir='-')
    limits_right.append(limit_right)
    limits_left.append(limit_left)

# Plotagem do gráfico da função
x = np.linspace(0, 10, 500)
y = np.tan(x)

plt.plot(x, y, label='f(x) = tan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) = tan(x)')
plt.legend()
plt.grid(True)
plt.show()

# Impressão dos limites
for i, point in enumerate(discontinuity_points):
    print('Descontinuidade em x =', point)
    print('Limite à direita:', limits_right[i])
    print('Limite à esquerda:', limits_left[i])
    print()
```

Função 2

a- Na função g(x) = 1/(x-5) se tem descontinuidade quando x = 5. Então, no intervalo de 0 a 10 tem descontinuidade em x = 5

```python

# Função g(x) = 1/(x-5)
x_sp = sp.symbols('x')
g_sp = 1 / (x_sp - 5)

# Ponto de descontinuidade
discontinuity_point = 5

# Cálculo dos limites à direita e à esquerda do ponto de descontinuidade
limit_right = sp.limit(g_sp, x_sp, discontinuity_point, dir='+')
limit_left = sp.limit(g_sp, x_sp, discontinuity_point, dir='-')

# Plotagem do gráfico da função
x = np.linspace(0, 10, 500)
y = 1 / (x - 5)

plt.plot(x, y, label='g(x) = 1/(x-5)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico da função g(x) = 1/(x-5)')
plt.legend()
plt.grid(True)
plt.show()

# Impressão dos limites
print('Descontinuidade em x =', discontinuity_point)
print('Limite à direita:', limit_right)
print('Limite à esquerda:', limit_left)
```

## Resposta Algébrica


Tem que encontrar primeiro a derivada da função

.

f(x) = ax^2 + bx + c

a=1   b=0   c=-1

x^n eh nx^(n-1)

derivada eh = f'(x)=2x

.

ponto (-1, 0):
f'(-1) = 2*(-1) = -2

ponto (0, -1):
f'(0) = 2*0 = 0

ponto (2, 3):
f'(2) = 2*2 = 4

## Resposta com código

```python
# Definição da função f(x) = x^2 - 1
def f(x):
    return x**2 - 1

# Ponto (-1, 0)
x1 = -1
y1 = f(x1)

# Ponto (0, -1)
x2 = 0
y2 = f(x2)

# Ponto (2, 3)
x3 = 2
y3 = f(x3)

# Cálculo da derivada da função
x_sp = sp.symbols('x')
f_sp = x_sp**2 - 1
f_prime_sp = sp.diff(f_sp, x_sp)

# Cálculo da inclinação da reta tangente nos pontos usando a derivada
slope1 = float(f_prime_sp.subs(x_sp, x1))
slope2 = float(f_prime_sp.subs(x_sp, x2))
slope3 = float(f_prime_sp.subs(x_sp, x3))

# Plotagem do gráfico da função
x = np.linspace(-3, 3, 100)
y = f(x)

plt.plot(x, y, label='f(x) = x^2 - 1')
plt.scatter(x1, y1, color='red', label='(-1, 0)')
plt.scatter(x2, y2, color='green', label='(0, -1)')
plt.scatter(x3, y3, color='blue', label='(2, 3)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) = x^2 - 1')
plt.legend()
plt.grid(True)
plt.show()

# Impressão da inclinação da reta tangente nos pontos
print('Inclinação da reta tangente no ponto (-1, 0):', slope1)
print('Inclinação da reta tangente no ponto (0, -1):', slope2)
print('Inclinação da reta tangente no ponto (2, 3):', slope3)
```

# Questão 5 (vale 1 ponto)
Você é um engenheiro de produção em uma fábrica de vidro que produz garrafas. A velocidade de produção das garrafas $V$ varia ao longo do tempo devido ao desgaste das máquinas e a fadiga dos trabalhadores. Depois de um mês de análise, você descobriu que a velocidade de produção é dada pela função $V(t) = 50 - 4t^2$ garrafas por hora, onde $t$ é o tempo em horas desde o início do turno de trabalho.
Seu gerente quer saber se há um momento específico durante o turno de trabalho em que a produção de garrafas é mais eficiente.
a) Como você poderia ajudar seu gerente a encontrar essa informação?
b) Se este momento existe, que momento é este?
c) Trace o gráfico da sua função usando apenas python, matplotlib, numpy e sympy e, caso este momento exista, marque este ponto no seu gráfico.

## Resposta  


a- Usando a função de velocidade de produção V(t) = 50 - 4t^2 para encontrar o valor max da produção

b- Resolver a equação V'(t) = 0 para encontrar o momento max


```python

# Função de velocidade de produção V(t)
t_sp = sp.symbols('t')
V_sp = 50 - 4 * t_sp**2

# Derivada da função de velocidade de produção
V_prime_sp = sp.diff(V_sp, t_sp)

# Resolver V'(t) = 0 para encontrar o momento em que a produção é mais eficiente
critical_points = sp.solve(V_prime_sp, t_sp)

# Verificar se existe um momento em que a produção é mais eficiente
if len(critical_points) > 0:
    efficient_time = critical_points[0]
    max_production = V_sp.subs(t_sp, efficient_time)
    print("O momento em que a produção é mais eficiente é t =", efficient_time)
    print("A produção máxima nesse momento é", max_production, "garrafas por hora.")
else:
    print("Não existe um momento específico em que a produção é mais eficiente.")

# Plotar o gráfico da função de velocidade de produção V(t)
t = np.linspace(0, 4, 100)
V = 50 - 4 * t**2

plt.plot(t, V, label='V(t) = 50 - 4t^2')
plt.xlabel('Tempo (horas)')
plt.ylabel('Produção de garrafas por hora')
plt.title('Velocidade de produção de garrafas')
plt.legend()
plt.grid(True)

# Marcar o ponto correspondente ao momento de máxima produção
if len(critical_points) > 0:
    plt.plot(efficient_time, max_production, 'ro', label='Máxima Produção')

plt.show()
```

# Questão 6 (vale 1 ponto)
Considere a função $f(x) = x^3 - 10x^2 + sin(x)$. Usando apenas o Python, numpy, sympy e matplotlib, trace o gráfico desta função e a reta tangente ao ponto $x=4$.

## Resposta

```python

# Função f(x) = x^3 - 10x^2 + sin(x)
def f(x):
    return x**3 - 10*x**2 + np.sin(x)

# Ponto para calcular a reta tangente
x_tangent = 4

# Definindo a função simbólica para cálculo da derivada
x_sp = sp.symbols('x')
f_sp = x_sp**3 - 10*x_sp**2 + sp.sin(x_sp)

# Cálculo da derivada
f_prime_sp = f_sp.diff(x_sp)

# Avaliando a função e a derivada no ponto dado
f_x = f(x_tangent)
f_prime_x = sp.N(f_prime_sp.subs(x_sp, x_tangent))

# Intervalo de x para o gráfico
x = np.linspace(-10, 10, 100)

# Valores de f(x) correspondentes
y = f(x)

# Plotagem do gráfico da função
plt.plot(x, y, label='f(x) = x^3 - 10x^2 + sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) = x^3 - 10x^2 + sin(x)')
plt.legend()
plt.grid(True)

# Cálculo da reta tangente
tangent_line = f_x + f_prime_x * (x - x_tangent)

# Plotagem da reta tangente
plt.plot(x, tangent_line, label='Reta tangente')
plt.legend()

plt.show()
```

#Questão 7 (vale 1 ponto)
Considere um fenômeno físico em que a posição de um objeto em movimento varia com o tempo. A taxa de variação da posição em relação ao tempo possui um componente trigonométrico e é dada pela função $f(t) = A * cos(t)$, onde $A$ é uma constante que determina a amplitude da variação. Sua tarefa é plotar o gráfico da função da taxa de variação e da função posição deste objeto considerando que $A$ é um número inteiro que varia entre 2 e 5.

## Resposta

```python

# Valores de A
A_values = np.arange(2, 6)

# Intervalo de tempo
t = np.linspace(0, 2*np.pi, 100)

# Plotagem dos gráficos
plt.figure(figsize=(10, 5))

# Plotagem do gráfico da taxa de variação
plt.subplot(1, 2, 1)
for A in A_values:
    f = A * np.cos(t)
    plt.plot(t, f, label=f'A = {A}')
plt.xlabel('Tempo')
plt.ylabel('Taxa de Variação')
plt.title('Gráfico da Taxa de Variação: f(t) = A * cos(t)')
plt.legend()
plt.grid(True)

# Plotagem do gráfico da posição
plt.subplot(1, 2, 2)
for A in A_values:
    position = A * np.sin(t)  # Calcula a posição integrando a taxa de variação
    plt.plot(t, position, label=f'A = {A}')
plt.xlabel('Tempo')
plt.ylabel('Posição')
plt.title('Gráfico da Posição: p(t) = ∫(A * cos(t)) dt')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
