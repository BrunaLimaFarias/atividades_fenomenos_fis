# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = np.linspace(0, 2, 100)
y1 = x * np.exp(-x**2)
y2 = x + 1

plt.plot(x, y1, label='y = xe^(-x^2)')
plt.plot(x, y2, label='y = x + 1')
plt.fill_between(x, y1, y2, where=(y1 >= y2), interpolate=True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre as curvas')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(2, color='red', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()


```
