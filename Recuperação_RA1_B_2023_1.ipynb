```python
## 1
def function(x):
    return (1 - np.cos(2*x)) / (x**2)

# Intervalo em torno de zero
interval = np.linspace(-0.1, 0.1, 1000)

# Valores para aproximação pela esquerda
x_left = np.linspace(-0.1, 0, 10)
y_left = function(x_left)

# Valores para aproximação pela direita
x_right = np.linspace(0, 0.1, 10)
y_right = function(x_right)

# Plot do gráfico da função
plt.plot(interval, function(interval), label='1 - cos(2x) / x^2')

# Plot dos pontos de aproximação
plt.plot(x_left, y_left, 'ro', label='Aproximação pela Esquerda')
plt.plot(x_right, y_right, 'bo', label='Aproximação pela Direita')

# Configurações do gráfico
plt.axvline(0, color='black', lw=0.5)  # Eixo vertical
plt.axhline(0, color='black', lw=0.5)  # Eixo horizontal
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Imprimir a tabela de aproximação
print('Tabela de Aproximação')
print('---------------------')
print('  x     |     y')
print('---------------------')
for i in range(len(x_left)):
    print(f'{x_left[i]:.4f}  |  {y_left[i]:.4f}')
for i in range(len(x_right)):
    print(f'{x_right[i]:.4f}  |  {y_right[i]:.4f}')

# Mostrar o gráfico
plt.title("1. lim x→0 1−cos(2x)/x^2")
plt.show()
```

# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, coloque o valor destes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico indicando as curvas correspondentes a cada equação. 

```python
def function(x):
    return np.where((-10 <= x) & (x < -3), np.sin(x),
                    np.where((-3 <= x) & (x < 0), np.cos(x),
                             np.where((0 <= x) & (x < 3), x**2,
                                      np.exp(x) - 10)))

# Intervalo de x
x = np.linspace(-10, 10, 1000)

# Valores correspondentes da função
y = function(x)

# Pontos de descontinuidade
descontinuidade = [-3, 0, 3]

# Plot do gráfico da função
plt.plot(x, y, label='f(x)')

# Marcar pontos de descontinuidade com círculos vermelhos
for d in descontinuidade:
    plt.plot(d, function(d), 'ro')

# Configurações do gráfico
plt.axhline(0, color='black', lw=0.5)  # Eixo horizontal
plt.axvline(0, color='black', lw=0.5)  # Eixo vertical
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
