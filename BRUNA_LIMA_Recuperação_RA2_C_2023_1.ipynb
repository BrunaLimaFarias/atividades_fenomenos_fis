# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 2.
$$h(x) = \frac{\ln(x)}{x}$$

## 1 - f(x)=x^3+5x^2−7x+9

uso a formula em cada um == d/dx (x^n) = nx^n-1

x^3  = 3x^2

5x^2 = 10x

-7x  = -7

9    = 0  (derivada de constante)

juntando tudo temos o resultado

## resultado == 3x^2 + 10x -7



```python
def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def f_der(x):
    return 3*x**2 + 10*x - 7

x = np.linspace(-10, 10, 400)
y = f(x)
y_der = f_der(x)

plt.plot(x, y, label='f(x)')
plt.plot(x, y_der, label="f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Função f(x) e derivada f'(x)")
plt.grid()
plt.show()
```

## 2 - h(x)=ln(x)/x

uso a regra do quociente d/dx (n/x)= xn'-nx'/x2

.

então as derivadas em relaçao a x

ln(x) = 1/x

x = 1

.

botando na formula

1/x.x-ln(x).1/x^2 = 1-ln(x)/x^2

.

## resultado == 1−ln(x)/x^2

```python
def h(x):
    return np.log(x) / x

def h_prime(x):
    return (1 - np.log(x)) / x**2

# Intervalo de x para o gráfico
x = np.linspace(0.1, 10, 400)

# Valores correspondentes de h(x) e h'(x)
y = h(x)
y_prime = h_prime(x)

# Plotando os gráficos
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid()
plt.title('Gráfico de h(x) = ln(x) / x')

plt.subplot(1, 2, 2)
plt.plot(x, y_prime)
plt.xlabel('x')
plt.ylabel("h'(x)")
plt.title("Gráfico da derivada h'(x)")

plt.tight_layout()
plt.grid()
plt.show()
```

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
def f(x):
    return x**2 - 98*x + 4

# intervalo de x
x = np.linspace(0, 100, 1000)

# valores de f(x)
y = f(x)

# indice do valor maximo
indice_max = np.argmax(y)

# obtendo o valor maximo e o valor correspondente do max
max_local = y[indice_max]
x_max_local = x[indice_max]

print("Maximo Local:")
print("f(x) =", max_local)
print("x =", x_max_local)
```
