#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import numpy as np
import sympy as sp
from sympy import symbols, solve
import matplotlib.pyplot as plt
```

```python

# Definindo as variáveis simbólicas
y = sp.symbols('y')

# Definindo as funções
x1 = y**2 - 2
x2 = y

# Encontrando os pontos de interseção
solutions = sp.solve(x1 - x2, y)
y_values = [float(sol) for sol in solutions]

# Limites para o eixo y
y_min = min(y_values) - 1
y_max = max(y_values) + 1

# Intervalo de valores para y
y_range = np.linspace(y_min, y_max, 100)

# Calculando os valores correspondentes de x1 e x2
x1_values = [sp.N(x1.subs(y, val)) for val in y_range]
x2_values = [sp.N(x2.subs(y, val)) for val in y_range]

# Convertendo para float
x1_values = [float(val) for val in x1_values]
x2_values = [float(val) for val in x2_values]

# Configurando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x1_values, y_range, label='x1 = y^2 - 2')
plt.plot(x2_values, y_range, label='x2 = y')
plt.fill_betweenx(y_range, x1_values, x2_values, where=(np.array(x1_values) >= np.array(x2_values)),color='lightblue')

# Configurando os limites dos eixos
plt.xlim(min(x1_values + x2_values) - 1, max(x1_values + x2_values) + 1)
plt.ylim(y_min, y_max)

# Adicionando legendas e título
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre as curvas x1 = y^2 - 2 e x2 = y')

# Exibindo o gráfico
plt.show()
```

#Questão 3 (vale até 2 pontos)
Um objeto está se movendo e experimenta diferentes acelerações em diferentes tempos. Os tempos e as respectivas acelerações são:

1. No instante $t = 1$ segundo, aceleração $a = 4 m/s²$
2. No instante $t = 3$ segundos, aceleração $a = -2 m/s²$
3. No instante $t = 4$ segundos, aceleração $a = 6 m/s²$
4. No instante $t = 6$ segundos, aceleração $a = 1 m/s²$
A posição inicial do objeto é $s_0 = 0 m$ e a velocidade inicial é $v_0 = 0 m/s$.

Você precisa imprimir com qualidade matemática (latex) a função encontrada para a aceleração, a função encontrada para a posição e a posição no instante $t=4.2s$. Além, é claro, dos gráficos de aceleração e posição.

Você poderá usar apenas as bibliotecas matplotlib, numpy e sympy.

## Resposta

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Dados fornecidos
t_values = [1, 3, 4, 6]  # Tempos (s)
a_values = [4, -2, 6, 1]  # Acelerações (m/s^2)
s0 = 0  # Posição inicial (m)
v0 = 0  # Velocidade inicial (m/s)

# Encontrando a função para aceleração
t = sp.Symbol('t')
a = sp.Piecewise((4, t < 1), (-2, t < 3), (6, t < 4), (1, True))

# Encontrando a função para posição
v = sp.integrate(a, t) + v0
s = sp.integrate(v, t) + s0

# Avaliando a posição no instante t=4.2s
t_eval = 4.2
s_eval = s.subs(t, t_eval)

# Imprimindo as funções encontradas
print('Função para aceleração:')
sp.pprint(a, use_unicode=True)
print('\nFunção para posição:')
sp.pprint(s, use_unicode=True)
print('\nPosição no instante t=4.2s:')
print(f's({t_eval}) = {s_eval}')

# Configurando os valores para plotagem dos gráficos
t_range = np.linspace(0, 7, 100)  # Faixa de tempo para os gráficos
a_func = sp.lambdify(t, a, modules='numpy')
s_func = sp.lambdify(t, s, modules='numpy')
a_values_plot = a_func(t_range)
s_values_plot = s_func(t_range)

# Plotando o gráfico de aceleração
plt.figure(figsize=(10, 5))
plt.plot(t_range, a_values_plot, label='Aceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.title('Gráfico de Aceleração')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o gráfico de posição
plt.figure(figsize=(10, 5))
plt.plot(t_range, s_values_plot, label='Posição')
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Gráfico de Posição')
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 5 (vale até 2 pontos)
Considere as seguintes curvas no plano cartesiano:

Curva A: $y = x^2$
Curva B: $y = 2x - 1$

Encontre os valores de [a, b], de modo que a área entre estas curvas neste intervalo seja igual a 6 unidades quadradas sabendo que $a=2$.

Usando o Python e apenas as bibliotecas matplotlib, numpy e sympy,
1. crie um código que plote estas duas curvas no intervalo entre $-10 \leq x \leq 10$;
2. crie um código que econtre o valor desejado de $b$;
3. crie um código que plote estas duas curvas sombreando em azul claro a área entre elas limitada pelo intervalo $[a,b] de forma que seja possível ver esta área completamente;
4. prove que o valor de b que você encontrou produzirá uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.


## Resposta

```python
# Etapa 1: Plotar as curvas A e B
x = np.linspace(-10, 10, 400)
y_a = x ** 2
y_b = 2 * x - 1

plt.plot(x, y_a, label='y = x^2')
plt.plot(x, y_b, label='y = 2x - 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas A e B')
plt.legend()
plt.grid(True)
plt.show()

# Etapa 2: Encontrar o valor de b
a = 2

x_sp = sp.symbols('x')
y_a_sp = x_sp ** 2
y_b_sp = 2 * x_sp - 1

b_sp = sp.symbols('b')
area = sp.integrate(y_b_sp - y_a_sp, (x_sp, a, b_sp))
equation = sp.Eq(area, 6)
solutions = sp.solve(equation, b_sp)

b = float(solutions[0])
print('Valor de b:', b)

# Etapa 3: Plotar as curvas sombreadas
fig, ax = plt.subplots()
ax.plot(x, y_a, label='y = x^2')
ax.plot(x, y_b, label='y = 2x - 1')

x_fill = np.linspace(a, b, 100)
y_fill_upper = np.minimum(2 * x_fill - 1, x_fill ** 2)
y_fill_lower = np.maximum(2 * x_fill - 1, x_fill ** 2)
ax.fill_between(x_fill, y_fill_lower, y_fill_upper, color='lightblue')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Área entre as curvas A e B')
ax.legend()
ax.grid(True)
plt.show()

# Etapa 4: Verificar a área com precisão
x_integrate = np.linspace(a, b, 1000)
y_integrate = 2 * x_integrate - 1 - x_integrate ** 2
area_approx = np.trapz(y_integrate, x_integrate)

print('Área aproximada:', area_approx)
```
